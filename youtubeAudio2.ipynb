## Import PyTube

```python
from pytube import YouTube
```

## Get the link to a YouTube Video

```python
%%html
"<iframe width="560" height="315" src="https://www.youtube.com/embed/CPGy11sEqAU?rel=0" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>"
```

## Create an YouTube Object

```python
yt = YouTube('"https://www.youtube.com/embed/CPGy11sEqAU')
```

## Get the YouTube Object Streams

```python
streams = yt.streams.filter(only_audio=True)
streams
```

## Get a Stream by itag

```python
stream = yt.streams.get_by_itag(140)
```

## Download the Stream

```python
stream.download()
```
