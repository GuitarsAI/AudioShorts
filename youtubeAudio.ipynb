## Import PyTube

```python
from pytube import YouTube
```

## Create an YouTube Object

```python
yt = YouTube('"https://www.youtube.com/embed/CPGy11sEqAU')
```

## Get the YouTube Object Streams

```python
streams = yt.streams.filter(only_audio=True)
streams
```

## Get a Stream by itag

```python
stream = yt.streams.get_by_itag(140)
```

## Download the Stream

```python
stream.download()
```

## Get Default Filename

## Install and Import FFMPEG Python

```python
pip install ffmpeg-python
```

```python
import ffmpeg
```

## Convert Audio to .mp3

```python
audio_stream = ffmpeg.input(stream.default_filename)
audio_stream = ffmpeg.output(audio_stream, 'converted.mp3', **{'b:a': '16k'})
ffmpeg.run(audio_stream)
```

## Play Converted Audio File

```python
Audio('converted.mp3')
```
