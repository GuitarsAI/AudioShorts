## Get Default Filename

## Install and Import FFMPEG Python

```python
pip install ffmpeg-python
```

```python
import ffmpeg
```

## Convert Audio to .mp3

```python
audio_stream = ffmpeg.input(stream.default_filename)
audio_stream = ffmpeg.output(audio_stream, 'converted.mp3', **{'b:a': '16k'})
ffmpeg.run(audio_stream)
```

## Play Converted Audio File

```python
Audio('converted.mp3')
```
