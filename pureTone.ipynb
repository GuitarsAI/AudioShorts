```python
from librosa import tone
```

```python
tone_440 = tone(440, duration=1, sr=44100)
```

```python
import IPython.display as ipd
```

```python
ipd.Audio(tone_440, rate=44100)
```

```python
import soundfile as sf
```

```python
sf.write('tone440_file.wav', tone_440, 44100, 'PCM_16')
```

```python
ipd.Audio("tone440_file.wav")
```

```python
from librosa import resample
```

```python
tone_440_16k = resample(tone_440, 44100, 16000)
```

```python
print("Tone_440",tone_440.shape)
print("Tone_440_16k",tone_440_16k.shape)
```

```python
ipd.Audio(tone_440_16k, rate=16000)
```
